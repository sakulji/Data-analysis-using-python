Extracting a table from a webpage

```python
forbes="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
```

```python
page1=requests.get(forbes)
```

```python
code2=BeautifulSoup(page1.text,'html')
print(code2)
```

```python
table=code2.find('table')
print(table)
```

```python
th=table.find_all('th')
print(th)
```

```python
heading=[i.text for i in th]
print(heading)
```

```python
import pandas as pd
```

```python
mydf=pd.DataFrame(columns=heading)
print(mydf)
```

```python
rows=table.find_all('tr')
print(rows)
```

```python
data=table.find_all('td')
print(data)
```

```python
data1=[i.text for i in data]
print(data1)
```

```python
df2=pd.DataFrame(columns=data1)
print(df2)
```

```python
ind=0
st=0
stop=4
for i in data1:
  if stop<=40:
    mydf.loc[ind]=data1[st:stop]
    st=st+4
    stop=stop+4
    ind=ind+1
```
