```python
import pandas as pd
df=pd.read_csv("/content/demodt.txt")
```

```python
df
```

```python

import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
#Decision tree algorithm
Literacy=int(input("Enter literacy rate:"))
Cleanliness=int(input("Enter clealiness rate:"))
Crime_Rate=int(input("Enter crime rate:"))
feature_col=['Literacy','Cleanliness','Crime_Rate']
feature=df[feature_col]#features
Target=df.Good #Target variable
DT=DecisionTreeClassifier()#creating the model
model=DT.fit(feature,Target)#fitting the model
print(model)
a=model.predict([[Literacy,Cleanliness,Crime_Rate]])
if a==1:
  print("Good state")
else:
  print("bad state")
```

```python
from sklearn.tree import export_graphviz,plot_tree
```

```python
plot_tree(model)#plot_tree(trained model name)
```

```python
#Random forest algorithm
from sklearn.ensemble import RandomForestClassifier  #loading the libraries
Literacy=int(input("Enter literacy rate:"))
Cleanliness=int(input("Enter clealiness rate:"))
Crime_Rate=int(input("Enter crime rate:"))
feature_col=['Literacy','Cleanliness','Crime_Rate']
feature=df[feature_col]#features
Target=df.Good #Target variable
RF=RandomForestClassifier()#creating the model
RF.fit(feature,Target)#fitting the model
print(model)
RF.predict([[Literacy,Cleanliness,Crime_Rate]])
if model==1:
  print("Good state")
else:
  print("bad state")
```

```python
#Saving all decision trees
#creating a directory
import os
output_dir="tree_visualisations"
os.makedirs(output_dir,exist_ok=True)
```

```python
#writing path
for i,tree in enumerate(RF.estimators_):
  tree_dot_file=os.path.join(output_dir,f"tree_{i}.dot")
  tree_png_file=os.path.join(output_dir,f"tree_{i}.png")
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file=tree_dot_file, feature_names=["Literacy", "Cleanliness", "Crime_Rate"],
                    class_names=[str(cls) for cls in RF.classes_], filled=True, rounded=True)
```

```python
command=f"dot -Tpng {tree_dot_file} -o {tree_png_file}"
os.system(command)
print(f"Tree {i} visualization saved to {tree_png_file}")
```
