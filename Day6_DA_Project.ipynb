```python
import pandas as pd
mydf=pd.read_csv("/content/Coimbra_breast_cancer_dataset.csv")
mydf
```

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
bmi=np.array(mydf['BMI']).reshape(-1,1)
insulin=np.array(mydf['Insulin'])
LR.fit(bmi,insulin)
LR.predict([[5.6]])
```

```python
#DECISION TREE
BMI=float(input("enter bmi"))
Glucose=int(input("enter glucose"))
Insulin=float(input("enter insulin"))
HOMA=float(input("enter homa"))
Leptin=float(input("enter leptin"))
Adiponectin=float(input("enter adiponectin"))
Resistin=float(input("enter resistin"))
MCP1=float(input("enter mcp"))
features_cols=['BMI','Glucose','Insulin','HOMA','Leptin','Adiponectin','Resistin','MCP.1']
features=mydf[features_cols]
target=mydf.Classification
DT1=DecisionTreeClassifier()
model=DT1.fit(features,target)
a=model.predict([[BMI,Glucose,Insulin,HOMA,Leptin,Adiponectin,Resistin,MCP1]])
if a==1:
  print("have cancer")
else:
  print("not having  cancer")
```

```python
from sklearn.tree import export_graphviz,plot_tree
```

```python
plot_tree(DT1)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
BMI=float(input("enter bmi"))
Glucose=int(input("enter glucose"))
Insulin=float(input("enter insulin"))
HOMA=float(input("enter homa i.e.used to quantify insulin resistance and beta-cell function"))
Leptin=float(input("enter leptin"))
Adiponectin=float(input("enter adiponectin"))
Resistin=float(input("enter resistin"))
MCP1=float(input("enter mcp"))
features_cols=['BMI','Glucose','Insulin','HOMA','Leptin','Adiponectin','Resistin','MCP.1']
features=mydf[features_cols]
target=mydf.Classification
RF1=RandomForestClassifier()
model=RF1.fit(features,target)
n_estimators=10
a=model.predict([[BMI,Glucose,Insulin,HOMA,Leptin,Adiponectin,Resistin,MCP1]])
if a==1:
  print("have cancer")
else:
  print("not having  cancer")
```

```python
import os
output_dir ="tree_visualizations_on_breast_cancer"
os.makedirs(output_dir,exist_ok=True)
```

```python
for i, tree in enumerate(model.estimators_):
  tree_dot_file=os.path.join(output_dir,f"tree_{i}.dot")
  tree_png_file=os.path.join(output_dir,f"tree_{i}.png")
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file=tree_dot_file, feature_names=['BMI','Glucose','Insulin','HOMA','Leptin','Adiponectin','Resistin','MCP.1'],
                    class_names=[str(cls) for cls in model.classes_], filled=True, rounded=True)
```

```python
command=f"dot -Tpng {tree_dot_file} -o {tree_png_file}"
os.system(command)
print(f"Tree {i} visualization saved to {tree_png_file}")
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
#Decision tree algorithm
ssc_p=float(input("Enter SSC percentage:"))
hsc_p=float(input("Enter HSC percentage:"))
degree_p=float(input("Enter degree percentage:"))
etest_p	=float(input("Enter entrance percentage:"))
mba_p=float(input("Enter MBA percentage:"))
feature_col=['ssc_p','hsc_p','degree_p','etest_p','mba_p']
feature=df[feature_col]#features
Target=df.status #Target variable
DT=DecisionTreeClassifier()#creating the model
model=DT.fit(feature,Target)#fitting the model
print(model)
a=model.predict([[ssc_p,hsc_p,degree_p,etest_p,mba_p]])
if a == "Placed":
  print("Yay! you are PLACED..")
else:
  print("Sorry you are NOT PLACED Better luck next time!")
```

```python
from sklearn.tree import export_graphviz,plot_tree
from sklearn.ensemble import RandomForestClassifier
```

```python
ssc_p=float(input("Enter SSC percentage:"))
hsc_p=float(input("Enter HSC percentage:"))
degree_p=float(input("Enter degree percentage:"))
etest_p	=float(input("Enter entrance percentage:"))
mba_p=float(input("Enter MBA percentage:"))
feature_col=['ssc_p','hsc_p','degree_p','etest_p','mba_p']
feature=df[feature_col]#features
Target=df.status #Target variable
RF=RandomForestClassifier()#creating the model
model=RF.fit(feature,Target)#fitting the model
n_estimators=10
print(model)
res=model.predict([[ssc_p,hsc_p,degree_p,etest_p,mba_p]])
if res=="Placed":
  print("you are placed")
else:
  print("you are not placed")
```

```python
command=f"dot -Tpng {tree_dot_file} -o {tree_png_file}"
os.system(command)
print(f"Tree {i} visualization saved to {tree_png_file}")
```
