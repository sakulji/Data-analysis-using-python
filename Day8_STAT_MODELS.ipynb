```python
#hypothesis testing
import numpy as np
from scipy import stats
```

```python
#loading the data
#two groups of data
data1 =np.array([28,21,26,29,23])
data2=np.array([21,27,25,28,19])
print(data1.mean())
print(data2.mean())
```

```python
#perform t_test assuming unequal variances
t_stat ,p_value=stats.ttest_ind(data1,data2,equal_var=False)
```

```python
#print the results
print("T-statistic:",t_stat)
print("P_value:",p_value)
```

```python
#alpha value
alpha=0.05
```

```python
#interpret the results
if p_value<alpha:
  print("reject null hypothesis.The means are significant")
else:
  print("Failed to reject the null hypothesis.The means are not significant")
```
